```python
#reading the dataset
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
#plotting single plots
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#labeling plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma v/s Virat Kohli Carrer')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#color(can also use hex code), line width and line style, marker and marker size
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashed',linewidth = 3,marker = 'o',markersize = 10)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dashdot',linewidth = 5,marker = 'D',markersize = 7)

plt.title('Rohit Sharma v/s Virat Kohli Carrer')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#line labeling
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashed',linewidth = 3,marker = 'o',markersize = 10,label = 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dashdot',linewidth = 5,marker = 'D',markersize = 7,label = 'Rohit')

plt.title('Rohit Sharma v/s Virat Kohli Carrer')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc = 'best')   #to finally print the line labels on the graph.
#we can change the value of loc where we wanna put the line labels box. the default value is best where it automatically finds the best position.
```

```python
#limiting axes(commonlyused to handle outliers)
price = [45000,48000,51000,50000,48000,46000,5000000]
year = [2016,2017,2018,2019,2020,2021,2022]
plt.plot(year, price)

plt.ylim(0,80000)
plt.xlim(2016,2020) 
```

```python
#grids
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashed',linewidth = 3,marker = 'o',markersize = 10,label = 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dashdot',linewidth = 5,marker = 'D',markersize = 7,label = 'Rohit')

plt.title('Rohit Sharma v/s Virat Kohli Carrer')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc = 'best')
plt.grid(True)
```

```python
#show()
plt.plot(batsman['index'],batsman['V Kohli'],color = 'red',linestyle = 'dashed',linewidth = 3,marker = 'o',markersize = 10,label = 'Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dashdot',linewidth = 5,marker = 'D',markersize = 7,label = 'Rohit')

plt.title('Rohit Sharma v/s Virat Kohli Carrer')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc = 'best')
plt.grid(True)

plt.show()
#we have to call this function to plot graph when we are not using a console like jupyter.
```

```python
#plotting a scatter plot
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
#scatter plot in pandas data
df = pd.read_csv('batter.csv')
df
```

```python
#plotting the graph on the top 50 batsman
df = df.head(50)
plt.scatter(df['avg'], df['strike_rate'])
```

```python
#labeling the data
df = df.head(50)
plt.scatter(df['avg'], df['strike_rate'],color='red',marker='1')
plt.title('Avg and Strike Rate Analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
#size
plt.scatter(tips['total_bill'],tips['tip'],s = tips['size']*20)
#each dot represents a single customer and bigger the dots represent that how many persons came together.
```

```python
#scatterplots using plt.plot
plt.plot(tips['total_bill'], tips['tip'], 'o')
#this is a faster techinique but we do get the less options here. used in extremely large datasets.
```

BAR CHART
used in:-
- bivariate analysis
- numerical v/s categorical data
- use case - aggregate analysis of groups

```python
#simple bar chat
children = [20,30,20,40,20]
colors = ['Red','Blue','Green','Yellow','Pink']

plt.bar(colors, children)
```

```python
#horizontal bar chart(used when no. of categories has increased)
plt.barh(colors, children)
```

```python
#color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
#overlapping bar charts
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
```

```python
#multiple bar plots
#there is no direct function in matplotlib to print multiple bar chats. we do it by adfjusting the width function.
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)
```

```python
#displaying the x axis with batsman names.
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
```

```python
#removing uneccesary comments
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```
